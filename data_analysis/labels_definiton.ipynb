# Labels Definition

```python
import pandas as pd
import numpy as np
import results
```

The labels are defined as follow:

 - Buy when the n next prices are higher than the current one
 - Sell when the n next prices are lower than the current one
 - Do nothing otherwise
 
 
 n = 24 in the following example

```python
df = pd.read_csv('../data/csv_files/btc_eur_3600.csv', index_col='timestamp')

window_length = 24
def local_extrema_label(window_array):
    if window_array[0] == max(window_array):
        return -1
    elif window_array[0] == min(window_array):
        return 1
    else:
        return 0

df['labels'] = df['closing_price'].rolling(window=window_length,center=False
                                ).apply(func=local_extrema_label
                                ).shift(-window_length+1)

#df[['labels','closing_price']]
```

```python
df['labels'].value_counts()
```

```python
labels = df['labels'].tolist()
prices = df['closing_price'].tolist()

transactions_number = 0
last_trade = 0
for label in labels:
    if label != last_trade and label != 0:
        last_trade = label
        transactions_number += 1

transactions_number
```
